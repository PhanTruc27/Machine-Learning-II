#**Estimate w(weights) and b(bias) by Gradient Descent**

```python
X.shape
```

```python
learning_rate = 0.1

def gradient_descent(X,y):
  W=0.0
  b=0.0
  N=X.shape[0]

  for i in range(100):
    error = y - (W*X + b)
    dw = -(1/N)*np.sum(error*X)
    db = -(1/N) *np.sum(error)

    W -= learning_rate * dw
    b -= learning_rate *db
  return b, W
```

```python
b,w = gradient_descent(X,y)
print (b, w)
```

```python
import numpy as np

learning_rate = 0.01

def gd(x, y, n_iter, tol=1e-5):
    w = 0.0
    b = 0.0
    N = len(x)

    for i in range(n_iter):
        # Tính sai số (dự đoán - thực tế)
        error = y - (w*x + b)

        # Tính đạo hàm (Gradient) theo w và b
        dw = -(1/N) * np.sum(error * x)
        db = -(1/N) * np.sum(error)

        # Cập nhật trọng số
        w -= learning_rate * dw
        b -= learning_rate * db

        # Early Stopping (Dừng sớm)
        # Nếu độ dốc quá nhỏ (gần bằng 0), nghĩa là đã tới đáy, thì dừng lại
        if np.abs(dw) < tol and np.abs(db) < tol:
            print(f"Hội tụ tại bước thứ: {i}")
            break

    return b, w
```
