```python
import numpy as np

learning_rate = 0.01

def gd(x, y, n_iter, tol=1e-5):
    w = 0.0
    b = 0.0
    N = len(x)

    for i in range(n_iter):
        # Tính sai số (dự đoán - thực tế)
        error = y - (w*x + b)

        # Tính đạo hàm (Gradient) theo w và b
        dw = -(1/N) * np.sum(error * x)
        db = -(1/N) * np.sum(error)

        # Cập nhật trọng số
        w -= learning_rate * dw
        b -= learning_rate * db

        # Early Stopping (Dừng sớm)
        # Nếu độ dốc quá nhỏ (gần bằng 0), nghĩa là đã tới đáy, thì dừng lại
        if np.abs(dw) < tol and np.abs(db) < tol:
            print(f"Hội tụ tại bước thứ: {i}")
            break

    return b, w
```

```python
import numpy as np
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
X,y = make_regression(n_samples = 100 , n_features= 2, noise =20, random_state=123)
```

```python
X.shape, y.shape
```

```python
 # Gradient Descent


def gradient_descent(X, y, learning_rate=0.01, n_iter=2000):
    N, d = X.shape
    w = np.zeros(d)
    b = 0.0

    for i in range(n_iter):
        y_hat = X @ w + b
        error = y - y_hat

        dw = -(2 / N) * X.T @ error
        db = -(2 / N) * np.sum(error)

        w -= learning_rate * dw
        b -= learning_rate * db

        if i % 200 == 0:
            loss = np.mean(error**2)
            print(f"iter {i}: loss={loss:.2f}")

    return w, b
```

```python
w_gd, b_gd = gradient_descent(X, y)
#Train the model with gradient descent
```

```python
# OLS estimation
X_ols = np.c_[np.ones(X.shape[0]), X]
beta = np.linalg.inv(X_ols.T @ X_ols) @ X_ols.T @ y

b_ols = beta[0]
w_ols = beta[1:]
```

```python
print("Gradient Descent:")
print("w:", w_gd)
print("b:", b_gd)

print("\nOLS:")
print("w:", w_ols)
print("b:", b_ols)
```
