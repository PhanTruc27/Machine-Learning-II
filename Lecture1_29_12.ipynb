

**1. Spurious Correlation (Common Trend)**
This occurs when two variables appear to be statistically related simply because they both trend upwards (or downwards) over time, rather than having a direct causal relationship. In econometrics, this is often caused by non-stationary data.


**2. Omitted Variable Bias (OVB)**
This arises when a model is too simple and excludes a relevant independent variable. Consequently, the variable that is included incorrectly "absorbs" the explanatory power of the missing factor, leading to biased results (often making the relationship look artificially stronger or "better" than it actually is).

# STAT MODEL

```python
import statsmodels.api as sm
```

```python
#x = sm.add_constant(df_scaled['VNM']) - learning in class
#y = df_scaled['GLD']
```

```python
df_scaled = df_scaled[['GLD', 'VNM']].dropna()

y = df_scaled['GLD']
x = sm.add_constant(df_scaled['VNM'])

model = sm.OLS(y, x)
results = model.fit()

print(results.summary())
```

**INTERPRET:**
The result indicates a strong positive association between VNM and GLD, rather than a causal relationship.

- The OLS results indicate a statistically significant positive linear association between gold prices (GLD) and the Vietnam Stock Index (VNM). The estimated coefficient on VNM is 0.859 (p < 0.001), implying that an increase in VNM is associated with an increase in GLD on average.

- The intercept is not statistically significant, which is not unusual given that both variables are standardized. The model exhibits a relatively high RÂ² of 0.738, suggesting that a substantial proportion of the variation in GLD is linearly associated with movements in VNM.

- The F-test confirms that the overall regression is statistically significant. However, the very low Durbinâ€“Watson statistic indicates strong positive autocorrelation in the residuals, which is common in financial time-series data and suggests that OLS inference should be interpreted with caution.

Overall, the results provide evidence of a strong positive linear relationship between GLD and VNM, although the findings should be interpreted as correlation rather than causation.

```python
plt.figure(figsize=(10, 6))
plt.scatter(df_scaled['VNM'], df_scaled['GLD'], alpha=0.5, label='Observed Data (Scaled)')
plt.plot(df_scaled['VNM'], results.predict(x), color='red', linewidth=2, label='OLS Regression Line')

plt.title('Linear Relationship: Vietnam ETF (VNM) predicting Gold (GLD)')
plt.xlabel('Standardized VNM Price')
plt.ylabel('Standardized GLD Price')
plt.legend()
plt.grid(True)
plt.show()
```

The scatter plot illustrates the standardized relationship between the Vietnam Stock Index (VNM) and gold prices (GLD). The upward-sloping OLS regression line confirms a positive linear relationship: as VNM increases, GLD tends to increase. Most data points cluster closely around the regression line, supporting the modelâ€™s high explanatory power (RÂ² = 0.738). Minor deviations reflect random variation or other factors not captured by VNM. Overall, the visualization reinforces the statistically significant positive association identified in the OLS regression.
